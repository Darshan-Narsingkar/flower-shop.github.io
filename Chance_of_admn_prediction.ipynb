```python
# step 1: import library
import pandas as pd
```

```python
# step 2: import data
admission = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
admission.head()
```

```python
admission.info()
```

```python
admission.describe()
```

```python
# step 3: Describe targer(y) and features (x)
admission.columns
```

```python
y = admission ['Chance of Admit ']
```

```python
x = admission[['Serial No', 'GRE Score', 'TOEFL Score', 'University Rating', ' SOP',
       'LOR ', 'CGPA', 'Research']]
```

```python
# step 4: train_test_split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7 ,random_state=2529)
```

```python
# check shape of train and test sample
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# step 5: select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6: train or fit model
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# step 7 : predict model
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
# step 8: model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error,mean_squared_error
```

```python
mean_absolute_error(y_test,y_predict)
```

```python
mean_absolute_percentage_error(y_test,y_predict)
```

```python
mean_squared_error(y_test,y_predict)
```
